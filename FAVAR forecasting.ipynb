```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
import os
```

```python
dataSet = pd.read_csv("C:\\Users\\Nikolay\\Desktop\\DataGermany.csv")
dataSet = dataSet.transpose()
new_header = dataSet.iloc[0] #grab the first row for the header
dataSet = dataSet[1:] #take the data less the header row
dataSet.columns = new_header #set the header row as the df header
```

```python
dataSet.fillna(dataSet.mean()) #replace na with mean value
dataSet = dataSet.iloc[::-1]
date_rng = pd.date_range(start='1/31/1991', end='9/30/2016', freq='M') #create monthly date
dataSet['date'] = date_rng
```

```python
dataSet.set_index('date', inplace=True)
```

```python
dataSet[['IP']].plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
os.getcwd()
```
